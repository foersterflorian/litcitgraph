```python
whitelist_data = pd.DataFrame()
for f in folder:
    df = pd.read_csv(f, sep=';', encoding='utf_8')
    assert (df.columns == target_columns).all()

    df = df.loc[df['Type']=='journal']
    df = df.dropna(subset=['SJR'])
    df['SJR'] = df['SJR'].apply(lambda x: x.replace(',', '.')).astype(float)
    df['SJR'] = (df['SJR'] * 1000).astype(int)
    df = df.sort_values('SJR', ascending=False)
    subset_data = df.iloc[:20]
    
    if whitelist_data.empty:
        whitelist_data = df.copy()
    else:
        whitelist_data = pd.concat([whitelist_data, subset_data])

whitelist_data = whitelist_data.drop_duplicates(subset=['Sourceid'])
```

## Using ISSN

```python
from collections.abc import Iterable

def extract_issn(entry):
    if ',' in entry:
        return [x.strip() for x in entry.split(',')]
    else:
        return entry

def flatten(x):
    for entry in x:
        if isinstance(entry, Iterable) and not isinstance(entry, (str, bytes)):
            yield from flatten(entry)
        else:
            yield entry
```

```python
issns = whitelist_data['Issn'].astype(str).apply(extract_issn).to_list()
whitelist_issns = frozenset(flatten(issns))
```
